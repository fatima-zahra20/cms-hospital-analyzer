```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("../cms_hospital.db")

# example query
query = """
SELECT 
    h.facility_id,
    h.facility_name,
    h.state,
    h.hospital_type,
    h.hospital_overall_rating
FROM hospital_general_info h
ORDER BY h.state, h.hospital_overall_rating DESC
LIMIT 20
"""

df = pd.read_sql_query(query, conn)
df
```
