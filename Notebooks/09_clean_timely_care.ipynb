```python
import pandas as pd

timely_care_df = pd.read_csv("../Data/cleaned/timely_effective_care_cleaned.csv")
print("Shape:", timely_care_df.shape)
print("\nColumns:")
for col in timely_care_df.columns:
    print(f"  {col}")
```

```python
print("unique measures:", timely_care_df["measure_id"].nunique())
print("\nunique conditions:")
print(timely_care_df["_condition"].value_counts())
```

```python
print(timely_care_df[["_condition", "measure_id", "measure_name"]].drop_duplicates().to_string())
```

```python
timely_care_df["score"] = timely_care_df["score"].replace("Not Available", None)
timely_care_df["score"] = pd.to_numeric(timely_care_df["score"], errors="coerce")

timely_care_df["sample"] = timely_care_df["sample"].replace("Not Available", None)
timely_care_df["sample"] = pd.to_numeric(timely_care_df["sample"], errors="coerce")

score_pivot = timely_care_df.pivot_table(
    index="facility_id",
    columns="measure_id",
    values="score",
    aggfunc="first"
).reset_index()
score_pivot.columns.name = None

hospital_info = timely_care_df[["facility_id", "facility_name", "address",
                                 "citytown", "state", "zip_code",
                                 "countyparish", "telephone_number"]].drop_duplicates("facility_id")

df_final = hospital_info.merge(score_pivot, on="facility_id", how="outer")

print("final shape:", df_final.shape)
print("duplicate facility_ids:", df_final["facility_id"].duplicated().sum())
```

```python
df_final.to_csv("../Data/cleaned/timely_effective_care_cleaned.csv", index=False)
print("timely_effective_care saved")
```
