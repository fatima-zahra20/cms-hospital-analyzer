```python
import pandas as pd

spending_df= pd.read_csv("../Data/cleaned/medicare_spending_cleaned.csv")
print("Shape:", spending_df.shape)
print("\nColumns:")
for col in spending_df.columns:
    print(f"  {col}")
```

```python
print("period values:")
print(spending_df["period"].value_counts())

print("\nclaim_type values:")
print(spending_df["claim_type"].value_counts())
```

```python
print("unique period + claim_type combinations:")
print(spending_df[["period", "claim_type"]].drop_duplicates().shape[0])

combo = spending_df.groupby(["period", "claim_type"]).size().reset_index(name="count")
print(combo.to_string())
```

```python
period_map = {
    "1 to 3 days Prior to Index Hospital Admission": "prior",
    "During Index Hospital Admission": "during",
    "1 through 30 days After Discharge from Index Hospital Admission": "after",
    "Complete Episode": "complete"
}

claim_map = {
    "Home Health Agency": "home_health",
    "Hospice": "hospice",
    "Inpatient": "inpatient",
    "Outpatient": "outpatient",
    "Skilled Nursing Facility": "snf",
    "Durable Medical Equipment": "dme",
    "Carrier": "carrier",
    "Total": "total"
}

spending_df["period_short"] = spending_df["period"].map(period_map)
spending_df["claim_short"] = spending_df["claim_type"].map(claim_map)
spending_df["combo_key"] = spending_df["period_short"] + "_" + spending_df["claim_short"]

print("sample combo keys:")
print(spending_df["combo_key"].unique())
```
