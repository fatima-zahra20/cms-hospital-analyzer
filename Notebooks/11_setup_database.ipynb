```python
import pandas as pd
import sqlite3
import os

conn = sqlite3.connect("../cms_hospital.db")

tables = {
    "hospital_general_info":   "../Data/cleaned/hospital_general_info_cleaned.csv",
    "readmissions_and_deaths": "../Data/cleaned/readmissions_and_deaths_cleaned.csv",
    "medicare_spending":       "../Data/cleaned/medicare_spending_cleaned.csv",
    "healthcare_infections":   "../Data/cleaned/healthcare_infections_cleaned.csv",
    "hcahps_patient_survey":   "../Data/cleaned/hcahps_patient_survey_cleaned.csv",
    "unplanned_visits":        "../Data/cleaned/unplanned_visits_cleaned.csv",
    "timely_effective_care":   "../Data/cleaned/timely_effective_care_cleaned.csv",
    "payment_and_value":       "../Data/cleaned/payment_and_value_cleaned.csv"
}

for table_name, filepath in tables.items():
    df = pd.read_csv(filepath)
    df.to_sql(table_name, conn, if_exists="replace", index=False)
    print(f"loaded {table_name} — {len(df)} rows ✓")

conn.close()
print("\ncms_hospital.db created!")
```
